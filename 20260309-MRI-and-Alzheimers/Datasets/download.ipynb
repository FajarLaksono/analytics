# Download or Update MRI and Alzheimers datasets

Link: https://www.kaggle.com/datasets/jboysen/mri-and-alzheimers

```python
import kagglehub
import shutil
import ipynbname
from pathlib import Path

# Target: the Datasets folder (same folder as this notebook)
target_dir = Path(ipynbname.path()).parent
print("Target directory:", target_dir)

# Download latest version to kaggle cache
cached_path = Path(kagglehub.dataset_download("jboysen/mri-and-alzheimers"))
print("Downloaded to cache:", cached_path)

# Copy all dataset files into this Datasets folder (overwrite if exists)
for src in cached_path.iterdir():
    dst = target_dir / src.name
    shutil.copy2(src, dst)
    print(f"Copied: {src.name}")

print("\nDone! Dataset files are now in:", target_dir)
```
